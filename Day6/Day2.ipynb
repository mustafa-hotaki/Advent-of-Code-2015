```
data = []
with open('inputs_day_2.txt', 'r') as f:
  for line in f:
    data.append(line.strip())

data[0]
```

```
# Parse
boxes = []
for x in data:
  box = tuple([int(d) for d in x.split('x')])
  boxes.append(box)
boxes[0]
```

## Part 1

```
total = 0
for l, w, h in boxes:
  sides = (l * w), (w * h), (h * l)
  surface_area = 2 * sum(sides)
  slack = min(sides)
  total += surface_area + slack

total
```

## Part 2

```
from functools import reduce
#boxes = [(2, 3, 4), (1, 1, 10)]
total = 0

for dimensions in boxes:
  wrap_length = 0
  for i, dim in enumerate(dimensions):
    if i != dimensions.index(max(dimensions)): # Only skips one if identical max
      wrap_length += dim * 2

  bow_length = reduce(lambda x, y: x * y, dimensions)
  
  total += (wrap_length + bow_length)

total
```
