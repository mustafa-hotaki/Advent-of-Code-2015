# Advent of Code 2015 - Day 8

```
data = []
with open("inputs_day_8.txt", "r") as f:
  for line in f:
    data.append(line.strip())
print('First 5 data points:', data[:5])
print('Number of data points:', len(data))
```

## Part 1

```
count_literals = 0
count_in_memory = 0

for string in data:
  count_literals += len(string)
  count_in_memory += len(eval(string))

count_literals - count_in_memory
```

## Part 2

```
count_literals = 0
count_new = 0

for string in data[:]:
  #print(string)
  count_literals += len(string)
  
  start = ['"', '\\', '"']
  end = ['\\', '"', '"']

  middle = []
  for c in list(string)[1:-1]:
    if c == '\\':
      middle.append(c)
      middle.append(c)
    elif c == '"':
      middle.append('\\')
      middle.append('"')

    else:
      middle.append(c)

  count_new += len(''.join(start + middle + end))

count_new - count_literals
```
