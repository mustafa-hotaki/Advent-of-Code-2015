# Advent of Code 2015 - Day 16

```
data = []
with open("inputs_day_16.txt", "r") as f:
  for line in f:
    data.append(line.strip())

print('First 2 data points:', data[:2])
print('Number of data points:', len(data))
```

```
# Parse
line = data[0]

aunts = {}
for line in data:
  aunt_num = int(line.split(' ')[1][:-1])
  attributes = {}
  for i, e in enumerate(line[line.index(':') + 2:].split(', ')):
    attributes[e.split(': ')[0]] = int(e.split(': ')[1])
  aunts[aunt_num] = attributes
  
aunts
```

```
requirements = {
  'children': 3,
  'cats': 7,
  'samoyeds': 2,
  'pomeranians': 3,
  'akitas': 0,
  'vizslas': 0,
  'goldfish': 5,
  'trees': 3,
  'cars': 2,
  'perfumes': 1
  }
```

```
# Find aunts that all of whose attributes match the requirements above
matches = {}
for aunt in aunts:

  aunt_attributes = aunts[aunt]

  attribute_match_counter = 0
  for key in aunt_attributes.keys():
    if aunt_attributes[key] == requirements[key]:
      attribute_match_counter += 1
    else:
      break

  if(attribute_match_counter == len(aunt_attributes)):
    matches[aunt] = attribute_match_counter


# Seems like there is only one match. If more existed, the answer would be the key with the larget value
matches
```

## Part 2

```
# Find aunts that all of whose attributes match the requirements above
matches = {}
for aunt in aunts:

  aunt_attributes = aunts[aunt]

  attribute_match_counter = 0
  for key in aunt_attributes.keys():
    if key == 'cats' or key == 'trees':
      if aunt_attributes[key] > requirements[key]:
        attribute_match_counter += 1
    elif key == 'pomeranians' or key == 'goldfish':
      if aunt_attributes[key] < requirements[key]:
        attribute_match_counter += 1
    else:
      if aunt_attributes[key] == requirements[key]:
        attribute_match_counter += 1

  if(attribute_match_counter == len(aunt_attributes)):
    matches[aunt] = attribute_match_counter


# Seems like there is only one match. If more existed, the answer would be the key with the larget value
matches
```
