# Advent of Code 2015 - Day

```
data = []
with open("inputs_day_5.txt", "r") as f:
  for line in f:
    data.append(line.strip())
print('First 5 data points:', data[:5])
print('Number of data points:', len(data))
```

## Part 1

```
def vowel_count(string):
  vowels = 'aeiou'
  counter = 0
  for v in vowels:
    counter += string.count(v)
  return counter
```

```
def two_adjacent_characters_are_same(string):
  for a, b in zip(string[:-1], string[1:]): # Loop over (a1, a2), (a2, a3), (a3, a4)...
    if a == b:
      return True
  return False
```

```
def forbidden_in_string(string, forbidden):
  for f in forbidden:
    if f in string:
      return True
  return False
```

```
nice_counter = 0

forbidden = ['ab', 'cd', 'pq', 'xy']

for string in data:
  nice_counter += vowel_count(string) >= 3 and two_adjacent_characters_are_same(string) and not forbidden_in_string(string, forbidden)
  
nice_counter
```

## Part 2

```
# Return true if strings contains a pair of any two letters that appears at least twice in the string without overlapping
def repeating_pair_at_least_twice(string):
  for i in range(len(string) - 1):
    pair = string[i : i + 2] # Grab two consecutive letters
    rest = string[i + 2 : ] # Only need to look forward
    if(pair in rest): # Does it appear again?
      return True
  return False
```

```
# Return true if string contains at least one letter which repeats with exactly one letter between them
import re
def repeating_letter_with_gap(string):
  characters = set(string)
  for c in characters:
    regex = c + '.' + c
    res = re.search(regex, string)
    if(res):
      return True
  return False
```

```
nice_counter = 0

for string in data:
  nice_counter += repeating_pair_at_least_twice(string) and repeating_letter_with_gap(string)

nice_counter
```
