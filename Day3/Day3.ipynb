```
data = []
with open('inputs_day_3.txt', 'r') as f:
  for line in f:
    data.append(line.strip())

data[0]
```

```
instructions = data[0]

house_map = {}

house_map = {(0, 0) : 1} # Start by having delievered to starting location

current_location = (0, 0)
for instruction in instructions:
  if instruction == '^':
    current_location = (current_location[0], current_location[1] + 1)
  elif instruction == 'v':
    current_location = (current_location[0], current_location[1] - 1)
  elif instruction == '>':
    current_location = (current_location[0] + 1, current_location[1])
  elif instruction == '<':
    current_location = (current_location[0] - 1, current_location[1])
  else:
    print('Oops')

  if(current_location in house_map):
    house_map[current_location] += 1
  else:
    house_map[current_location] = 1
```

```
instructions = data[0]

house_map = {}
house_map = {(0, 0) : 2} # Start by having delievered to starting location

# Santa
current_location = (0, 0)
for instruction in instructions[::2]: # Loop over instructions at indexes 0, 2, 4, 6, ...
  if instruction == '^':
    current_location = (current_location[0], current_location[1] + 1)
  elif instruction == 'v':
    current_location = (current_location[0], current_location[1] - 1)
  elif instruction == '>':
    current_location = (current_location[0] + 1, current_location[1])
  elif instruction == '<':
    current_location = (current_location[0] - 1, current_location[1])
  else:
    print('Oops')

  if(current_location in house_map):
    house_map[current_location] += 1
  else:
    house_map[current_location] = 1

# Robo-Santa
current_location = (0, 0) # Reset
for instruction in instructions[1::2]: # Loop over instructions at indexes 1, 3, 5, 7, ...
  if instruction == '^':
    current_location = (current_location[0], current_location[1] + 1)
  elif instruction == 'v':
    current_location = (current_location[0], current_location[1] - 1)
  elif instruction == '>':
    current_location = (current_location[0] + 1, current_location[1])
  elif instruction == '<':
    current_location = (current_location[0] - 1, current_location[1])
  else:
    print('Oops')

  if(current_location in house_map):
    house_map[current_location] += 1
  else:
    house_map[current_location] = 1
```
