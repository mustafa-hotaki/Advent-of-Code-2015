```
data = []
with open('input.txt', 'r') as f:
  data = f.readlines()
```

## Part 1

```
instructions = data[0]
floor = 0
for i in instructions:
  floor += 1 if i == '(' else -1
floor
```

## Part 2

```
floor = 0
for c, i in enumerate(instructions):
  floor += 1 if i == '(' else -1
  if floor == -1:
    print(c + 1)
    break
```
